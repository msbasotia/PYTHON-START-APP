```python
start = datetime.datetime.now()
file = pd.read_csv("practice_pandas.csv")
end = datetime.datetime.now()
```

```python
start1 = datetime.datetime.now()
chunk = pd.read_csv("practice_pandas.csv",chunksize=1000)
end1 = datetime.datetime.now()
file1 = pd.concat(chunk)
```

```python
print(f"time taken to read csv = {(end1-start1).total_seconds()} sec")
```

# Creating New Columns

## Simple Addition of columns

```python
file["pg_custom"] = file["point_granularity"] + 2
```

```python
file.head()
```

## Basis a condition

```python
custom_const = []
for val in file["construction"]:
    if val == "Wood":
        custom_const.append("Wood Palm")
    else:
        custom_const.append(val)

file["custom_const"] = custom_const

# https://www.dezyre.com/recipes/insert-new-column-based-on-condition-in-python
```
